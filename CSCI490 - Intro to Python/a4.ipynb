# Part 1 Read Data

```python
food_prices = open("food-prices-weekly.txt", "r")
```

```python
file_iter = iter(food_prices)
names = next(file_iter)

category_1 = names[0:10]
category_2 = names[12:26]
category_3 = names[28:55]
category_4 = names[57:66]
category_5 = names[68:77]

strip_1 = category_1.strip()
strip_2 = category_2.strip()
strip_3 = category_3.strip()
strip_4 = category_4.strip()
strip_5 = category_5.strip()

category_list = [strip_1,strip_2,strip_3,strip_4,strip_5]
```

```python
# create list
data = []
```

```python
for line in file_iter:
    slice_1 = line[0:10]
    slice_2 = line[12:26]
    slice_3 = line[28:55]
    slice_4 = line[57:66]
    slice_5 = line[68:77]

    strip_1 = slice_1.strip()
    strip_2 = slice_2.strip()
    strip_3 = slice_3.strip()
    strip_4 = slice_4.strip()
    strip_5 = slice_5.strip()

    new_list = [strip_1,strip_2,strip_3,strip_4,strip_5]

    new_dict = dict(zip(category_list,new_list))

    data.append(new_dict)

data
    
```

# Part 2

```python
for diction in data:
    full_date = diction.get("Date")
    yyyy_mm = full_date[0:7]
    diction['Month'] = yyyy_mm

data
```

# Part 3

```python
monthly_data = []
monthly_calc = []
```

```python
monthly_totals = []
j = 1
for it in data:
    found = False
    state = it.get('State')
    cat = it.get('Category')
    month = it.get('Month')
    dollars = it.get('Dollars')
    if j != 1: #prevents the first one getting stuck in a loop
        for i in monthly_calc:
            check_state = i.get('State')
            check_cat = i.get('Category')
            check_month = i.get('Month')
            if state == check_state and cat == check_cat and month == check_month:
                found = True
                old_total = i.get('Total')
                new_total = int(old_total) + int(dollars)
                carrier = {'Total': new_total}
                i.update(carrier)
                old_count = i.get('Count')
                new_count = old_count + 1
                carrier = {'Count': new_count}
                i.update(carrier)
        
    if found == False:
        new_dict = {'State': state, 'Category': cat, 'Month': month, 'Total': dollars, 'Count': 1}
        monthly_calc.append(new_dict)
    
    j += 1

# I know this cell takes a very long time. I thought about doing it another way, but because I wanted to check
# that the month, category, and state matched before adding it to the total, this would prevent a mishap if the
# file was not in order. There's definitely faster solutions banking on the fact that the file is in order.
# And probably faster ones that do the same thing as this but I could not think of something

# I understand if I lose points for this. 

# It took an average of 1.5 min to run this cell

```

```python
for it in monthly_calc:
    month = it.get('Month')
    cat = it.get('Category')
    state = it.get('State')

    total = it.get('Total')
    count = it.get('Count')

    avg = int(total) / count

    new_dict = {'Month': month, 'State': state, 'Category': cat, 'DollarsAvg': avg}
    monthly_data.append(new_dict)

monthly_data
```

# Part 4

```python
key_list = []
for keys in monthly_data[0]:
    key_list.append(keys)

key_list
```

```python
with open('food-prices-monthly.txt', 'w') as f:
    f.write(f'{key_list[0]:8}' + f'{key_list[1]:15}' + f'{key_list[2]:28}' + f'{key_list[3]:12}' + '\n')
    for it in monthly_data:
        month = it.get('Month')
        state = it.get('State')
        cat = it.get('Category')
        avg = it.get('DollarsAvg')
        avg = round(avg,1)
        f.write(f'{month:8}' + f'{state:15}' + f'{cat:28}' + f'{avg:12}' + '\n')
```
